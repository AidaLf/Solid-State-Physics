```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python

def omegaP(c1,c2, kpiS):
  return (c1+c2+(c1**2+c2**2+2*c1*c2*np.cos(kpiS))**0.5)**0.5

def omegaN(c1,c2,kpiM):
  return (c1+c2-(c1**2+c2**2+2*c1*c2*np.cos(kpiM))**0.5)**0.5

def plotOmega(c1,c2):
    fig=plt.subplots(figsize=(10,7))
    plt.grid()
    kpiM=np.linspace(-math.pi,math.pi); 
    arr1=np.linspace(-2*math.pi,-math.pi); arr2=np.linspace(math.pi,2*math.pi)
    kpiS=np.concatenate((arr1,arr2))
    plt.plot(kpiS[kpiS<-math.pi], omegaP(c1,c2,kpiS)[kpiS<-math.pi], c='darkorange',linewidth=3, label='Optical mode');plt.plot(kpiS[kpiS>math.pi], omegaP(c1,c2,kpiS)[kpiS>math.pi],c='darkorange',linewidth=3);
    # plt.plot(kpiM, omegaP(c1,c2,kpiM),"--",c='mediumblue')
    plt.plot(kpiM[kpiM<math.pi+1], omegaN(c1,c2,kpiM)[kpiM<math.pi+1], c='mediumblue', linewidth=3, label='Accoustical mode');
    import importlib
    importlib.reload(plt)
    theta=np.arange(-2*math.pi,2*math.pi+math.pi/2,step=(math.pi/2))
    labels=['-2π', '-3π/2', '-π', '-π/2', '0', 'π/2', 'π', '3π/2', '2π']
    plt.xticks(theta, labels)
    # plt.plot(kpiS[kpiS<-math.pi], omegaN(c1,c2,kpiS)[kpiS<-math.pi],"--", c='orange');plt.plot(kpiS[kpiS>math.pi], omegaN(c1,c2,kpiS)[kpiS>math.pi],"--",c='orange');
    plt.title("ω(k) vs ka for a diatomic chain with equal masses\n"+"$C_{2}$="+str(c2/c1)+"$C_{1}$", fontsize=16)
    plt.xlabel("ka", fontsize=14); plt.ylabel("ω(k)", fontsize=14);plt.xticks=()
    plt.legend()
    pi=math.pi
    plt.show()


plotOmega(1,1)
plotOmega(1,2)
plotOmega(1,10)
plotOmega(1,100)
```
