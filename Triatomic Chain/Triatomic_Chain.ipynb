```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
N=100; pi=math.pi
fig=plt.subplots(figsize=(12,8))
plt.grid(alpha=0.5)

#for m=1, c=1

#coeefs:
a=3;b=-13;c=27/2
omega1=np.empty(N); omega2=np.empty(N); omega3=np.empty(N)

ka1=np.linspace(0,pi,N)
for i in range(len(ka1)):
  coeff=[a,b,c,-4*np.sin(ka1[i]/2)**2]
  omegaArr=np.sqrt(np.roots(coeff))
  omega1[i]=omegaArr[2]
plt.plot(ka1,omega1,lw=3, label="Accoustical Mode",c="crimson")
plt.plot(-ka1,omega1,lw=3, c="crimson")


ka2=np.linspace(pi,2*pi,N)
for i in range(len(ka2)):
  coeff=[a,b,c,-4*np.sin(ka2[i]/2)**2]
  omegaArr=np.sqrt(np.roots(coeff))
  omega2[i]=omegaArr[1]
plt.plot(ka2,omega2,lw=3,label="Optical Mode 1", c="darkorange")
plt.plot(-ka2,omega2,lw=3, c="darkorange")


ka3=np.linspace(2*pi,3*pi,N)
for i in range(len(ka3)):
  coeff=[a,b,c,-4*np.sin(ka3[i]/2)**2]
  omegaArr=np.sqrt(np.roots(coeff))
  omega3[i]=omegaArr[0]
plt.plot(ka3,omega3,lw=3,label= "Optical Mode 2", c="navy")
plt.plot(-ka3,omega3,lw=3, c="navy")


plt.plot(np.full((10),pi),np.linspace(0,2.5,10),'k--', alpha=0.5); plt.plot(np.full((10),-pi),np.linspace(0,2.5,10),'k--', alpha=0.5);
plt.plot(np.full((10),2*pi),np.linspace(0,2.5,10),'k--', alpha=0.5);plt.plot(np.full((10),-2*pi),np.linspace(0,2.5,10),'k--', alpha=0.5)

theta=np.arange(-3*math.pi,3*math.pi+math.pi/2,step=(math.pi)); labels=[ '-3π', '-2π', '-π', '0', 'π', '2π', '3π']; plt.xticks(theta, labels);
plt.legend()
plt.xlabel("ka", fontsize=16); plt.ylabel("$\\frac{ω(k)}{\\sqrt{\\frac{2c}{m}}}$", fontsize=20)
plt.title("$\\frac{ω(k)}{\\sqrt{\\frac{2c}{m}}}$ vs $ka$ for a triatomic chain with three different masses\n $m_{1}=m$,  $m_{2}=\\frac{3}{4}m$,  $m_{3}=\\frac{1}{2}m$ ", fontsize=16)
plt.ylim(0,1.8)

plt.show()
```

```python
ka=np.linspace(-3*math.pi,3*math.pi,N)

for i in range(len(ka)):
  coeff=[a,b,c,-4*(np.sin(ka[i]/2))**2]
  omegaArr=np.sqrt(np.roots(coeff))
  omega1[i]=omegaArr[0]
  omega2[i]=omegaArr[1]
  omega3[i]=omegaArr[2]


fig=plt.subplots(figsize=(12,8))
plt.grid(alpha=0.5)


plt.plot(ka,omega1,lw=3,label= "Optical Mode 1", c="navy"); plt.plot(ka,omega2,lw=3,label="Optical Mode 2", c="darkorange"); plt.plot(ka,omega3,lw=3, label="Accoustical Mode",c="crimson")
plt.plot(np.full((10),pi),np.linspace(0,2.5,10),'k--', alpha=0.5); plt.plot(np.full((10),-pi),np.linspace(0,2.5,10),'k--', alpha=0.5);
plt.plot(np.full((10),2*pi),np.linspace(0,2.5,10),'k--', alpha=0.5);plt.plot(np.full((10),-2*pi),np.linspace(0,2.5,10),'k--', alpha=0.5)
plt.legend()
theta=np.arange(-3*math.pi,3*math.pi+math.pi/2,step=(math.pi)); labels=[ '-3π', '-2π', '-π', '0', 'π', '2π', '3π']; plt.xticks(theta, labels);
plt.xlabel("ka", fontsize=16); plt.ylabel("$ω(k)$", fontsize=16)
plt.title("$ω(k)$ vs $ka$ for a triatomic chain with three different masses\n $m_{1}=m$,  $m_{2}=\\frac{3}{4}m$,  $m_{3}=\\frac{1}{2}m$ ", fontsize=16)
plt.show()
```
