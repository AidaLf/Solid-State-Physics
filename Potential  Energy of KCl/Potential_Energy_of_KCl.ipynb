```python
from mpl_toolkits.axisartist.axislines import AxesZero
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
e=1
lam=2134.5962 #in ev
Rho=0.326
K=14.4 #in ev A
def UC(alpha,r):
  return -(alpha*K*e**2)/(r)
def UR(z,r):
  return z*lam*math.exp(-r/Rho)
def U(r,alpha,z):
  return UC(alpha,r)+UR(z,r)
```

```python
#For NaCl
num=1000
R=np.linspace(1,6,num)

alpha_NaCl=1.748; Z_NaCl=6;
alpha_ZnS=1.64; Z_ZnS=4; 

UR_N=np.empty([num]); UC_N=np.empty([num]);
UC_Z=np.empty([num]); UR_Z=np.empty([num])
for i in range(len(R)):
  UC_N[i]=UC(alpha_NaCl, R[i])
  UR_N[i]=UR(6,R[i])
  UC_Z[i]=UC(alpha_ZnS, R[i])
  UR_Z[i]=UR(4,R[i])


fig, ax = plt.subplots(figsize=(8, 5))

ax.grid(True, linestyle='-.')
ax.tick_params( labelsize='medium', width=3)

ax.plot(R,UC_N, label='Coulomb Energy')
ax.plot(R,UR_N, label='Repulsive Energy')
ax.plot(R,UC_N+UR_N, label='Total Energy')
ax.legend()
ax.set_title('Energy per molecule of KCl crystal in NaCl structure')
ax.set_xlabel('Distance in Å')
ax.set_ylabel('Energy in eV')
ax.set_ylim(-20,20)

x = R[np.argmin(UC_N+UR_N)]
y = np.min(UC_N+UR_N)
plt.scatter(x,y,None,'black')
txt1='Equilibrium position'+'\n'+ 'R= '+str(round(x,3))+"Å"
plt.annotate(txt1, xy=(x, y), xytext=(40, 40),
              textcoords='offset points', arrowprops=dict(arrowstyle='->', color='black'), ha='center', va='top')

plt.show()



```

```python

fig, ax = plt.subplots(figsize=(8, 5))
ax.grid(True, linestyle='-.')
ax.tick_params( labelsize='medium', width=3)
x_EP = R[np.argmin(UC_Z+UR_Z)]
y_EP = np.min(UC_Z+UR_Z)
plt.scatter(x_EP, y_EP)
ax.plot(R,UC_Z, label='Coulomb Energy')
ax.plot(R,UR_Z, label='Repulsive Energy')
ax.plot(R,UC_Z+UR_Z, label='Total Energy')
ax.legend()
ax.set_title('Energy per molecule of KCl crystal in ZnS structure')
ax.set_xlabel('Distance in Å')
ax.set_ylabel('Energy in eV')
ax.set_ylim(-20,20)
txt2='Equilibrium position'+'\n'+ 'R= '+str(round(x_EP,3))+"Å"
plt.scatter(x_EP,y_EP,None,'black')
plt.annotate(txt2, xy=(x_EP, y_EP), xytext=(40, 40),
              textcoords='offset points', arrowprops=dict(arrowstyle='->', color='black'), ha='center', va='top')
plt.show()
```

```python

fig, ax = plt.subplots(figsize=(8, 5))
ax.grid(True, linestyle='-.')
ax.tick_params( labelsize='medium', width=3)
ax.plot(R,UC_N+UR_N, label='NaCl Structure', color='mediumblue')
ax.plot(R,UC_Z+UR_Z, label='ZnS Structure',  color='m')
ax.legend()
ax.set_title('Energy per molecule of KCl crystal')
ax.set_xlabel('Distance in Å')
ax.set_ylabel('Energy in eV')
ax.set_ylim(-12,25)
x1 = R[np.argmin(UC_Z+UR_Z)]
y1 = np.min(UC_Z+UR_Z)
x2 = R[np.argmin(UC_N+UR_N)]
y2 = np.min(UC_N+UR_N)
plt.scatter(x1, y1, color='m')
plt.scatter(x2, y2, color='mediumblue')
plt.show()
```
